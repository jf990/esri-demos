```python
import datetime

%matplotlib inline
import matplotlib.pyplot as pd
from IPython.display import display, YouTubeVideo

import arcgis
from arcgis.gis import GIS
from arcgis.features.analyze_patterns import interpolate_points
from arcgis.geocoding import geocode

gis = GIS('https://www.arcgis.com', 'HermanMunster')
```

```python
my_map = gis.map("Point Dume state beach", 12)
```

```python
my_map
```

```python
trails = gis.content.get("2e4b3df6ba4b44969a3bc9827de746b3")
my_map.add_layer(trails)

temp = gis.content.get("b8df6517ceac42af9ab483089296ed04")
my_map.add_layer(temp)
```

```python
# interpolated_temp_item = interpolate_points(temp, output_name='avg_temp', field='Mean_T_c_06_Jun')
interpolated_temp_item = gis.content.get("50c4db90311d4211a072f889fb078b40")
my_map.add_layer(interpolated_temp_item, options={'opacity': 0.4})
```
